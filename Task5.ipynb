# Viewing the regional profit using barchart

The below data shows that the *maximum* profit is in the **West** region and *minimum* in the **South**

```
objects = ('East', 'West', 'South', 'Central')
y_pos = np.arange(len(objects))
performance = [91522.8,108418,46749.4,39706.4]
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Profit')
plt.title('Regional Profit')

plt.show()
```

```
plt.rcParams['figure.figsize'] = (18, 7)
color = plt.cm.copper(np.linspace(0, 1, 40))
df['Sub-Category'].value_counts().plot.bar(color = color)
plt.title('Most Popular items', fontsize = 24)
plt.xticks(rotation = 90, fontsize = 15)
plt.show()
```

The above Plot shows that the most common item in all the regions is Binders and Paper
